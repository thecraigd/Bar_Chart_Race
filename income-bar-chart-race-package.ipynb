```python
# import our libraries
import pandas as pd
import matplotlib.pyplot as plt
import bar_chart_race as bcr


# create a DataFrame from the CSV we sourced from the www.gapfinder.org
income = pd.read_csv('income_per_person_gdppercapita_ppp_inflation_adjusted.csv', index_col=0)


# drop the future projections from the dataset
income.drop(income.iloc[:,-20:], axis=1, inplace=True)

# pivot the DF to get it into the correct format as per https://pypi.org/project/bar-chart-race/
income_new = income.T # could have also used income.transpose() here but apparently this does the same thing?!?

display(income_new.head())
```

```python
bcr.bar_chart_race(
    df=income_new,
    filename='income.mp4',
    period_length=250,
    title='GDP per capita (in 2011 USD)',
    n_bars=10)
```
